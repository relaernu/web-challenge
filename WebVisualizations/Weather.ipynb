```python
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')

from datetime import datetime
```

```python
df = pd.read_csv("resources/cities.csv")
```

```python
df
```

```python
cols = df.columns.tolist()
cols[4] = "Timestamp"
df.columns = cols
```

```python
df["Date"] = df["Timestamp"].apply(lambda x : datetime.fromtimestamp(x).date())
df
```

```python
date = df["Date"].drop_duplicates()
strdate = date[0].strftime('%d/%m/%Y')
strdate = "Date:" + strdate
```

```python
# plot Max temp vs Lat
ax = df.plot.scatter(x="Lat", y="Max Temp")
# ax.set_title("Max temperature VS Latitude")
ax.set_xlabel("Latitude")
ax.set_ylabel("Max Temperature(°F)")
x = min(ax.get_xticks())
y = max(ax.get_yticks())
ax.text(x, y, strdate)
```

```python
# plot Wind Speed to Latitude
ax = df.plot.scatter(x="Lat", y="Wind Speed")
#ax.set_title("Wind speed VS Latitude")
ax.set_xlabel("Latitude")
ax.set_ylabel("Wind speed(mph)")
x = min(ax.get_xticks())
y = max(ax.get_yticks())
ax.text(x,y,strdate)
```
